## Dropna() function - handling the missing values by dropping it

```python
net = pd.read_csv('netflix-dataset.csv')
```

```python
net
```

```python
net.isnull().sum()
```

```python
net.shape
```

```python
# dropna() - function drop the entire row or col having null value
net.dropna().shape
```

```python
# axis parameter in dropna() function drop the col value
# without this parameter, function drop row having null value
net.dropna(axis=1).shape
```

```python
# how='' parameter in dropna() function drop the row having 
# all - drop the row having all null values
# any - drop the row having any one null values
net.dropna(how='all').shape
```

```python
# subset parameter in dropna() function drop the cols if they have null values
net.dropna(subset=['Director','Release_Date']).shape
```

## Fillna() - function fill the places of missing values

```python
# this will fill all missing all empty spaces with 0
# either the missing place is integer or a string
# which is not appropriate for strings

net.fillna(0)
```

```python
# to fill strings and integers apartly
# inplace - function makes the changes permanent in the file

net['Director'].fillna("Not specified", inplace=True)
net['Cast'].fillna("Not Mentioned", inplace=True)
net['Release_Date'].fillna("00-00-0000", inplace=True)
```

```python
# ffill - fills the missing place by copying the previous/above place value
# bfill - fills the missing place by copying the next place value
net['Country'].fillna("Not Mentioned", inplace=True)
net['Rating'].fillna("Not Rated", inplace=True)
```
