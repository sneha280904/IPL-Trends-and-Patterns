```python
def run_scored(batsman_name):
    batter = data2[data2['batsman'] == batsman_name]
    return (batter.groupby('bowling_team')['batsman_runs'].sum().sort_values(ascending = False)).head(3)
```

```python
run_scored('MS Dhoni')
```

```python
run_scored('V Kohli')
```

```python
d_over = data2[data2['over'] > 15]
d_over
```

```python
all_bats = d_over.groupby('batsman')['batsman_runs'].count()
x = all_bats > 200
all_bats[x]
```

```python
# tolist() - function converts series into list
batsman_list = all_bats[x].index.tolist()
batsman_list
```

```python
final = data2[data2['batsman'].isin(batsman_list)]
runs = final.groupby('batsman')['batsman_runs'].sum()
balls = final.groupby('batsman')['batsman_runs'].count()
```

```python
sr = (runs/balls)*100
```

```python
sr.sort_values(ascending=False)
```
