```python
def run_scored(batsman_name):
    batter = data2[data2['batsman'] == batsman_name]
    return (batter.groupby('bowling_team')['batsman_runs'].sum().sort_values(ascending = False)).head(3)
```

```python
run_scored('MS Dhoni')
```

```python
run_scored('V Kohli')
```

```python
run_scored('RG Sharma')
```

```python
d_over = data2[data2['over'] > 15]
d_over
```

```python
all_bats = d_over.groupby('batsman')['batsman_runs'].count()
x = all_bats > 200
all_bats[x]
```

```python
# tolist() - function converts series into list
batsman_list = all_bats[x].index.tolist()
batsman_list
```

```python
final = data2[data2['batsman'].isin(batsman_list)]
runs = final.groupby('batsman')['batsman_runs'].sum()
balls = final.groupby('batsman')['batsman_runs'].count()
```

```python
sr = (runs/balls)*100
```

```python
sr.sort_values(ascending=False)
```

```python
def team_cityground(city, team):
    m1 = data[data['city'] == city]
    m2 = m1['team1'] == team
    m3 = m1['team2'] == team
    return m1[m2 | m3]

cmcc = team_cityground('Chennai', 'Chennai Super Kings')
cmcc['winner'].value_counts()
```

```python
bmbc = team_cityground('Bangalore', 'Royal Challengers Bangalore')
```

```python
bmbc['winner'].value_counts()
```

```python
team_hometowns = {
    'Chennai Super Kings': 'Chennai',
    'Mumbai Indians': 'Mumbai',
    'Kolkata Knight Riders': 'Kolkata',
    'Royal Challengers Bangalore': 'Bangalore',
    'Delhi Daredevils': 'Delhi',
    'Sunrisers Hyderabad': 'Hyderabad',
    'Kings XI Punjab': 'Chandigarh',
    'Rajasthan Royals': 'Jaipur'
}
```

```python
keys_list = list(team_hometowns.keys())
values_list = list(team_hometowns.values())
```

```python
for i in range (0,8):
    team = team_cityground(values_list[i], keys_list[i])
    own_homeground = team[team['winner'] == keys_list[i]]
    print(own_homeground['winner'].value_counts())
```
